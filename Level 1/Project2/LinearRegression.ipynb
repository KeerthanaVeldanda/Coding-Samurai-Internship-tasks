```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('advertising_spend_vs_sales_revenue.csv')
```

```python
df
```

```python
plt.scatter(df['Advertising Spend ($)'],df['Sales Revenue ($)'])
plt.xlabel('Advertising Spend (in $)')
plt.ylabel('Sales (in $)')
plt.title('Sales vs Advertising Spend')
plt.show()
```

```python
x=df[['Advertising Spend ($)']]
y=df['Sales Revenue ($)']
xtrain,xtest,ytrain,ytest=train_test_split(x,y,random_state=10)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
pre=model.predict(xtest)
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (b):", model.intercept_)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(ytest,pre)
r2 = r2_score(ytest,pre)
print(f"Mean Squared Error:{mse:.2f}")
print(f"R^2 Score:{r2:.2f}")
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, model.predict(x), color='blue')
plt.xlabel('Advertising Spend ($)')
plt.ylabel('Sales Revenue ($)')
plt.title('Linear Regression:')
plt.show()
```
